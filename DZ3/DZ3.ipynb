```python
import pandas as pd
import numpy
sales=pd.read_csv('Electronic_sales.csv')
#sales.head(15)

#Задание 1

method_of_pay=sales.groupby(['Customer ID','Payment Method']).size().reset_index(name='Count')#предпочитаемый метод оплаты
#method_of_pay=sales.groupby(['Customer ID','Payment Method']).agg({'Payment Method': 'count'})
#method_of_pay.head(50)
expenses=sales.groupby('Customer ID')['Total Price'].sum()# общие траты
#expenses.head()
Dop_expenses=sales.groupby('Customer ID')['Add-on Total'].sum()# Доп траты
Dop_expenses.head()


```

```python
#Задание 2
sales['Purchase Date'] = pd.to_datetime(sales['Purchase Date']) 
sales['Month']=sales['Purchase Date'].dt.month
sales['Quarter']=sales['Purchase Date'].dt.quarter
#sales.head(15)

Sh_sales=sales.groupby('Shipping Type')['Total Price'].sum()
Sh_sales=pd.DataFrame(Sh_sales,columns=['Total Price'])
#Sh_sales.head()

Pt_sales=sales.groupby('Product Type')['Total Price'].sum()
Pt_sales=pd.DataFrame(Pt_sales,columns=['Total Price'])
#Pt_sales.head()

Ms_sales=sales.groupby('Month')['Total Price'].sum()
Ms_sales=pd.DataFrame(Ms_sales,columns=['Total Price'])
#Ms_sales.head()

Qs_sales=sales.groupby('Quarter')['Total Price'].sum()
Qs_sales=pd.DataFrame(Qs_sales,columns=['Total Price'])
Qs_sales.head()
```

```python
import seaborn as sns
sns.histplot(data=Qs_sales, x='Quarter')
```
